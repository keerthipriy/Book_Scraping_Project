### Web scraping on Multiple pages:-

```python
import requests
page=1
while page <=4:
    url=f"https://books.toscrape.com/catalogue/category/books/sequential-art_5/page-{page}.html"
    res=requests.get(url)
    
    from bs4 import BeautifulSoup
    soup=BeautifulSoup(res.text,"html.parser")
    title=soup.find_all("h3")
    price=soup.find_all("p",class_="price_color")
    for t,p in zip(title,price):
        book_title=t.find("a")["title"]
        book_price=p.text.strip()
        print(f"title:{book_title}*** price:{book_price}")
    page=page+1
```

## Data Cleaning

```python
import pandas as pd

df = pd.DataFrame(all_books)
df.head()
```

```python
df
```

```python
df.isnull().sum()   
```

```python
df = df.dropna(subset=['title', 'price'])   
df = df.fillna({'rating': 'Not Rated'})     
```

```python
df['title'] = df['title'].str.strip()       # Remove leading/trailing spaces
df['title'] = df['title'].str.title()       # Optional: make consistent title case
```

```python
df = df.drop_duplicates(subset='title', keep='first')
```

```python
print(df.info())
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for clean visuals
sns.set(style="whitegrid")
```

```python
import matplotlib.pyplot as plt

# Simple bar graph for books vs price
plt.figure(figsize=(10,6))
plt.bar(df['title'], df['price'])
plt.xticks(rotation=75, ha='right')
plt.xlabel('Book Title')
plt.ylabel('Price (£)')
plt.title('Books vs Price')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Clean and convert 'price' column to float
df['price'] = (
    df['price']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)  # Remove everything except digits and dot
    .astype(float)
)

# Top 10 most expensive books
top = df.nlargest(10, 'price')

# Column chart
plt.figure(figsize=(10,6))
plt.bar(top['title'], top['price'], color='skyblue')
plt.xticks(rotation=75, ha='right')
plt.ylabel('Price (£)')
plt.title('Top 10 Most Expensive Books')
plt.tight_layout()
plt.show()
```
